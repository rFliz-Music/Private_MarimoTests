```python
import numpy as np

# --------------------------------------------------
# Chord representation
# --------------------------------------------------

def major(root):
    return tuple(sorted([
        root % 12,
        (root + 4) % 12,
        (root + 7) % 12
    ]))

def minor(root):
    return tuple(sorted([
        root % 12,
        (root + 3) % 12,
        (root + 7) % 12
    ]))

# Cube Dance vertices:
# 12 major + 12 minor triads
vertices = []

for r in range(12):
    vertices.append(major(r))

for r in range(12):
    vertices.append(minor(r))

# --------------------------------------------------
# Indexing
# --------------------------------------------------

idx_of = {ch:i for i,ch in enumerate(vertices)}
chord_of = {i:ch for ch,i in idx_of.items()}

N = len(vertices)

A = np.zeros((N, N), dtype=np.uint8)

# --------------------------------------------------
# P10 generator
# --------------------------------------------------

def p10_neighbors(chord):

    chord = set(chord)

    for note in chord:

        for delta in (-1, +1):

            moved = (note + delta) % 12

            if moved in chord:
                continue

            new_chord = tuple(
                sorted((chord - {note}) | {moved})
            )

            yield new_chord

# --------------------------------------------------
# Build graph
# --------------------------------------------------

for chord in vertices:

    i = idx_of[chord]

    for nbr in p10_neighbors(chord):

        if nbr not in idx_of:
            continue

        j = idx_of[nbr]

        A[i, j] = 1
        A[j, i] = 1

print(A.shape)


# --------------------------------------------------
# Name Render (for readability)
# --------------------------------------------------

NOTE_NAMES = [
    "C", "C#", "D", "Eb", "E", "F",
    "F#", "G", "Ab", "A", "Bb", "B"
]

CHORD_NAMES = {}

# Major triads
for r in range(12):

    triad = tuple(sorted([
        r,
        (r+4)%12,
        (r+7)%12
    ]))

    CHORD_NAMES[triad] = f"{NOTE_NAMES[r]}M"

# Minor triads
for r in range(12):

    triad = tuple(sorted([
        r,
        (r+3)%12,
        (r+7)%12
    ]))

    CHORD_NAMES[triad] = f"{NOTE_NAMES[r]}m"

# Augmented triads
for r in range(4):

    triad = tuple(sorted([
        r,
        (r+4)%12,
        (r+8)%12
    ]))

    CHORD_NAMES[triad] = f"{NOTE_NAMES[r]}+"
```

```python
cube_neighbors = [
    x for x in p10_neighbors(chord_of[0])
    if x in CHORD_NAMES
]

[CHORD_NAMES[x] for x in cube_neighbors]


# neighborood = list(p10_neighbors(chord_of[0]))

# [CHORD_NAMES[x] for x in neighborood]
```
